```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def derivada_progresivaOh2(f,x,h):
    
    f_prima=(1/(2*h))*(-3*f(x)+4*f(x+h)-f(x+2*h))
    
    return f_prima
```

```python
def derivada_centralOh2(f,x,h):
    
    f_prima=(f(x+h)-f(x-h))/(2*h)
    
    return f_prima
```

```python
x=np.linspace(0.1,1.1,101)
#h=0.01
h=x[1]-x[0]
x
```

```python
def funcion(x):
    
    f_x=np.sqrt(np.tan(x))
    
    return f_x
```

```python
dProgresivaO2=derivada_progresivaOh2(funcion,x,h)
```

```python
dCentralO2=derivada_centralOh2(funcion,x,h)
```

```python
def dfuncion(x):
    
    f_prima=(1)/((np.cos(2*x)+1)*(np.sqrt(np.tan(x))))
    
    return f_prima
```

```python
derivadaAnalitica=dfuncion(x)
plt.plot(x,derivadaAnalitica,label="Derivada Analitica")
plt.plot(x,dProgresivaO2,label="Derivada Progresiva")
plt.plot(x,dCentralO2,label="Derivada Central")
plt.legend()
```

```python
errorNodalProg=np.abs(derivadaAnalitica-dProgresivaO2)
errorNodalCent=np.abs(derivadaAnalitica-dCentralO2)
plt.plot(x,errorNodalProg,label="Error Nodal Progresivo")
plt.plot(x,errorNodalCent,label="Error Nodal Central")
plt.legend()
```

Al tener $h=0.01 \implies \mathcal{O}(h^2)=0.001$, por lo tanto, podemos afirmar que, si bien no son igual de exactas, ambas aproximaciones tienen la misma presición de $\mathcal{O}(h^2)$.
