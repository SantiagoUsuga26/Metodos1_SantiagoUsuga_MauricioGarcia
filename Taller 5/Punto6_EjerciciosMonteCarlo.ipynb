```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
```

```python
def CreateSphere(N, R):
    
    X = np.zeros(N)
    Y = np.zeros_like(X)
    Z = np.zeros_like(X)
    
    
    for i in tqdm(range(N)):
        
        u = np.random.rand()
        ro = R*u**(1/2)
        
        theta = np.random.uniform(0,2*np.pi)
        phi = np.random.uniform(0,np.pi)
        
        X[i] = ro*np.sin(phi)*np.cos(theta)
        Y[i] = ro*np.sin(phi)*np.sin(theta)
        Z[i] = ro*np.cos(phi)
    
    
    return X,Y,Z
```

```python
X,Y,Z = CreateSphere( int(1e6), 1 )
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, projection = '3d')

#configurar límites de los ejes
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-1, 1)

ax.view_init(10, 60)

ax.scatter(X,Y,Z,color='b')
```

```python
def f(r,phi):
    func = np.exp(r)*r**2*np.sin(phi)
    return  func
```

```python
def GetPoints(R,phimin,phimax):
    u = np.random.rand()
    ro = R*u**(1/2)
    phi = np.random.uniform(phimin,phimax)

    
    return ro,phi
```

```python
N = int(1e6)
Sample = np.zeros(N)

for i in range(N):
    ro,phi = GetPoints(1,0,np.pi)
    Sample[i] = f(ro,phi)
    
```

```python
Integral = 4*np.pi*np.average( Sample )
```
