```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=1 #número de personas a calcular
```

```python
def casosFavorables(n):
    r=((365-n)/365)
    return r
```

```python
probabilidad=1
dist=[]
for i in range(1,81):
    n=i
    probabilidad=probabilidad*casosFavorables(n)#Probabilidad conjunta de que no cumplan el mismo día
    dist.append(1-probabilidad)
```

```python
x=np.linspace(1,81,80)
plt.title("Probabilidad en función del número de personas")
plt.ylabel("P(n)")
plt.xlabel("n")
plt.plot(x,dist)
```
